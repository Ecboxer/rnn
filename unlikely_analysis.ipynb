```python
import pandas as pd
import numpy as np
import nltk
```

```python
df_act = df.groupby('actual').sum().sort_values('freq', ascending=False)
df_act.head()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
```

```python
def clean_actual(x):
    if x not in stop_words and len(x) > 1:
        try:
            return tokenizer.tokenize(x.lower())[0]
        except IndexError:
            return np.nan
    else:
        return np.nan
```

```python
df['actual_clean'] = df['actual'].apply(lambda x: clean_actual(x))
df.head()
```

```python
len(df['actual_clean'].unique())
```

```python
df_act_clean = df.groupby('actual_clean').sum().sort_values('freq', ascending=False)
df_act_clean.head()
```
